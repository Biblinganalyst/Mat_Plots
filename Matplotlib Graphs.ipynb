# Scatter Plot

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[1,3,5,2,4,6]
z=[1,4,3,6,5,2]

colors=[10,29,15,27,32,48]
sizes=[400,200,300,500,600,100]


# aplha is used for the transparency of data ponts and it has the value between 0 and 1
# cmap is used to represent the data point on the line bar
#sizes are used for the different shapes of data point

plt.scatter(x,y, c=colors, s=sizes, cmap="viridis")   
plt.scatter(x,z, c="r", alpha=0.5)

#colorbar is used to plot te colormap on the right side of the plot
t=plt.colorbar()
t.set_label("colorbar", fontsize=15)


plt.xlabel("X", fontsize=15)
plt.ylabel("y", fontsize=15)
plt.title("Scatter plot", fontsize=15)

plt.show
```

# Bar plot

# Multiple Bar Plot

```python
import numpy as np
import matplotlib.pyplot as plt

X = ['Group A','Group B','Group C','Group D']
Ygirls = [10,20,20,40]
Zboys = [20,30,25,30]

X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, Ygirls, 0.4, label = 'Girls')
plt.bar(X_axis + 0.2, Zboys, 0.4, label = 'Boys')

plt.xticks(X_axis, X)
plt.xlabel("Groups")
plt.ylabel("Number of Students")
plt.title("Number of Students in each group")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 3
ind = np.arange(N)
width = 0.25

xvals = [8, 9, 2]
bar1 = plt.bar(ind, xvals, width, color = 'r')

yvals = [10, 20, 30]
bar2 = plt.bar(ind+width, yvals, width, color='g')

zvals = [11, 12, 13]
bar3 = plt.bar(ind+width*2, zvals, width, color = 'b')

plt.xlabel("Dates")
plt.ylabel('Scores')
plt.title("Players Score")

plt.xticks(ind+width,['2021Feb01', '2021Feb02', '2021Feb03'])
plt.legend( (bar1, bar2, bar3), ('Player1', 'Player2', 'Player3') )
plt.show()
```

# Stem Plot

```python
from matplotlib import pyplot as plt

x=[1,2,3,4,5,6]
y=[9,2,4,2,7,3]

#linefmt is used to give the shape to the plot, bottom will start from zero, basefmt is used to change the color of base, 
#use_line_collection is used to give the differet to different plots  
plt.stem(x,y, linefmt="--", markerfmt="ro", bottom=0, basefmt="g", label="Python", orientation="horizontal", 
         use_line_collection=False)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Stem Plot")

plt.legend()
plt.show()
```

# BOX plot

# Area VS Stack Plot

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6,7]
area1=[7,6,5,4,3,2,1]
area2=[7,1,5,3,4,2,6]
area3=[1,3,2,4,2,6,3]

l=["area1", "area2", "area3"]

#plt.stackplot(x, area1, area2, area3, labels=l)
plt.stackplot(x, area1, area2, area3, labels=l, colors=["r", "g", "m"], baseline="zero")  #change colors of area
#baseline= sym, zero, wiggle

plt.legend()
plt.show()
```

# STEP Plot

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,2,4,5,1]

plt.step(x,y, color="r", marker="o", ms=10, mfc="g", label="Python")  #color change of line, marker, marker size, marker color, label

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Py")

plt.legend(loc=2)
plt.show()
```

# Fill Between Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([1,3,2,5,4])
plt.plot(x,y, color="r")
plt.fill_between(x,y, color="g", where=(x>2) & (x<4), alpha=0.3)

plt.title("Python")
plt.xlabel("x_axis")
plt.ylabel("y_axis")

plt.show()
```
